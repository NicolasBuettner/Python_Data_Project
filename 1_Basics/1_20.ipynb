# 1_20_4
Use the requests library to fetch data from the URL 'https://pokeapi.co/api/v2/pokemon/pikachu' and print the JSON response.

```python
import requests

url = 'https://pokeapi.co/api/v2/pokemon/pikachu'

response = requests.get(url)
data = response.json()
print(data)
```
